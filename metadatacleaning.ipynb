```python
import pandas as pd
import ast

df = pd.read_csv("bot_metadata7.csv", dtype=str)

def extract_social_keys(x):
    if not isinstance(x, str):
        return ""
    try:
        data = ast.literal_eval(x)    # convert string dictionary to real dict
    except:
        return ""
    if not isinstance(data, dict):
        return ""
    keys = []
    for k in data.keys():
        name = str(k).strip()
        if name != "":
            keys.append(name)
    return ",".join(keys)

# create the correct column
df["social_combined"] = df["socials"].apply(extract_social_keys)

# preview
print(df[["bot_id", "social_combined"]].head())
```

```python
import pandas as pd
import re
import ast

# df must already contain:
# bot_id, name, short_description, long_description, prefix, invite_url, permissions, socials

def extract_commands(short_desc, long_desc, prefix):
    text = f"{short_desc} {long_desc} {prefix}"
    text = text.replace("\n", " ").replace("\r", " ")
    text = " ".join(text.split())

    patterns = [
        r"![a-zA-Z0-9_]+",
        r"/[a-zA-Z0-9_]+",
        r"\.[a-zA-Z0-9_]+",
        r"\?[a-zA-Z0-9_]+",
        r"\$[a-zA-Z0-9_]+"
    ]

    cmds = set()
    for p in patterns:
        for m in re.findall(p, text):
            cmds.add(m)

    if isinstance(prefix, str) and prefix.strip() != "":
        px = prefix.strip()
        for m in re.findall(rf"{re.escape(px)}[a-zA-Z0-9_]+", text):
            cmds.add(m)

    return ", ".join(sorted(cmds))


def extract_github(socials_raw):
    """Return a single GitHub URL string, or empty string."""
    if not isinstance(socials_raw, str) or socials_raw.strip() == "":
        return ""

    # try to parse dict-like string
    data = None
    try:
        data = ast.literal_eval(socials_raw)
    except Exception:
        data = None

    # Case 1: socials is a dict
    if isinstance(data, dict):
        # direct key "GitHub"
        gh_val = data.get("GitHub")
        if isinstance(gh_val, str) and "github.com" in gh_val.lower():
            return gh_val.strip()
        if isinstance(gh_val, (list, tuple)) and gh_val:
            # take first GitHub-looking string
            for v in gh_val:
                if isinstance(v, str) and "github.com" in v.lower():
                    return v.strip()
        # or search all dict values
        for v in data.values():
            if isinstance(v, str) and "github.com" in v.lower():
                return v.strip()

    # Case 2: fallback to regex on raw string
    m = re.search(r"https?://github\.com/[^\s'\"}]+", socials_raw)
    if m:
        return m.group(0).strip()

    return ""


# 1. Extract commands
df["commands"] = df.apply(
    lambda r: extract_commands(
        r.get("short_description", ""),
        r.get("long_description", ""),
        r.get("prefix", "")
    ),
    axis=1
)

# 2. Extract a single GitHub URL per bot
df["github_url"] = df["socials"].apply(extract_github)

# 3. Filter bots with REAL values and a non-empty GitHub URL
df_filtered = df[
    df["name"].notna() &
    (df["name"].astype(str).str.strip() != "") &
    df["invite_url"].notna() &
    (df["invite_url"].astype(str).str.strip() != "") &
    df["commands"].notna() &
    (df["commands"].astype(str).str.strip() != "") &
    df["permissions"].notna() &
    (df["permissions"].astype(str).str.strip() != "") &
    (df["permissions"].astype(str).str.strip() != "0") &
    df["github_url"].notna() &
    (df["github_url"].astype(str).str.strip() != "") &
    (df["github_url"].astype(str).str.contains("github.com", case=False))
]

# 4. Build final CSV for your dynamic analysis (only bots with a single GitHub URL field)
commands_df = df_filtered[[
    "bot_id",
    "name",
    "prefix",
    "invite_url",
    "permissions",
    "commands",
    "github_url"
]]

# 5. Save
commands_df.to_csv("bot_commands_final_github_only.csv", index=False)
```
